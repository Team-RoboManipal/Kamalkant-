```python
#feature scaling
features = df.drop(columns=['RainTomorrow'])

# Z-score normalization for each feature column
for column in features.columns:
    mean = features[column].mean()
    std = features[column].std()
    df[column] = (features[column] - mean) / std
```

```python
df=df.sample(frac=1, random_state=42).reset_index(drop=True)

train_size = int(0.8 * len(df))
df_train,df_test=df[:train_size],df[train_size:]
y_train,y_test=df_train['RainTomorrow'].to_numpy(),df_test['RainTomorrow'].to_numpy()
X_train,X_test=df_train.drop("RainTomorrow",axis=1).to_numpy(),df_test.drop("RainTomorrow",axis=1).to_numpy()

```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def loss(y, y_pred):
    return -np.mean(y*np.log(y_pred) + (1-y)*np.log(1-y_pred))
```

```python
def predict(X, weights,b):
    f_wb = np.dot(X, weights)+b
    if sigmoid(f_wb) >= 0.5:
        return 1
    else:
        return 0
```

```python
def compute_cost(X,y,w,b):
    m=X.shape[0]
    cost=0
    for i in range(m):
        z=np.dot(X[i],w)+b
        f_wb=sigmoid(z)
        cost+=(-1*y[i])*np.log(f_wb)-(1-y[i])*np.log(1-f_wb)
    cost/=m
    return cost
```

```python
def compute_gradient(X,y,w,b):
    m,n=X.shape
    dj_dw=np.zeros(n)
    dj_db=0
    for i in range(m):
        f_wb=sigmoid(np.dot(X[i],w)+b)
        err=f_wb-y[i]
        for j in range(n):
            dj_dw[j]+=err*X[i,j]
        dj_db+=err

    dj_dw/=m
    dj_db/=m
    return dj_dw,dj_db
```

```python
from IPython.display import display, clear_output
def gradient_descent(X,y,w_in,b_in,lr,epoch):
    w=w_in
    b=b_in
    LOSS_hist=[]
    for i in range(epoch):
        dj_dw,dj_db=compute_gradient(X,y,w,b)
        w-=lr*dj_dw
        b-=lr*dj_db
        LOSS_hist.append(loss(y,sigmoid(np.dot(X,w)+b)))
        clear_output(wait=True)
        display(f'Epoch {i + 1}/{epoch} LOSS: {loss(y,sigmoid(np.dot(X,w)+b))}')

    return w,b,LOSS_hist
```

```python
w,b,LOSS_hist=gradient_descent(X_train,y_train,np.zeros(X_train.shape[1]),0,0.1,2000)
```

```python
plt.plot(range(len(LOSS_hist)), LOSS_hist)
plt.xlabel("Iterations")
plt.ylabel("LOSS")
plt.title("Iteration vs LOSS")
plt.grid(True)
plt.show()
```

```python
y_pred = [predict(X, w, b) for X in X_test]
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```
