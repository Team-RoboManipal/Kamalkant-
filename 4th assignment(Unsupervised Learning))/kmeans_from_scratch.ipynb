```python
import numpy as np
import matplotlib.pyplot as plt

class KMeans:
    def __init__(self,k,max_iteration=200):
        self.k=k
        self.centroids=None
        self.max_iteration=max_iteration
        self.wcss=0 #Within-Cluster Sum of Squares
        self.num_iter=0
    def calculate_wcss(self, X, y):
        """Calculate the Within-Cluster Sum of Squares."""
        wcss = 0
        for i in range(self.k):
            cluster_points = X[y == i]
            if len(cluster_points) > 0:
                wcss += np.sum((cluster_points - self.centroids[i])**2)
        return wcss

    @staticmethod
    def dist(data_point,centroids):
        return np.sqrt(np.sum((centroids-data_point)**2,axis=1))

    def initialize_centroids(self, X):
        """K-Means++ initialization."""
        n_samples = X.shape[0]
        centroids = []

        # Randomly select the first centroid
        centroids.append(X[np.random.choice(range(n_samples))])

        # Select the remaining k-1 centroids
        for _ in range(1, self.k):
            distances = np.min([np.linalg.norm(X - c, axis=1)**2 for c in centroids], axis=0)
            probabilities = distances / np.sum(distances)
            next_centroid = X[np.random.choice(range(n_samples), p=probabilities)]
            centroids.append(next_centroid)

        return np.array(centroids)

    def fit(self,X):
        self.centroids=self.initialize_centroids(X)

        for _ in range(self.max_iteration):
            self.num_iter+=1
            y=[]
            for data_point in X:
                distances=KMeans.dist(data_point,self.centroids) # finds distance b/w centroid and datapoint
                cluster_num=np.argmin(distances)
                y.append(cluster_num) #appends the index of min distance

            y=np.array(y)

            cluster_indices=[]

            for i in range(self.k):
                cluster_indices.append(np.argwhere(y==i) ) # groups index based in category

            cluster_centers=[]

            for i , indices in enumerate(cluster_indices):
                if len(indices)==0:
                    cluster_centers.append(self.centroids[i])
                else:
                    cluster_centers.append(np.mean(X[indices],axis=0)[0])

            cluster_centers=np.array(cluster_centers)

            if np.max(self.centroids-cluster_centers) < 0.0001:
                break
            else:
                self.centroids=cluster_centers

        self.wcss=self.calculate_wcss(X,y)
        return y
```

```python

# Range of k values to test
k_values = range(1, 10)
wcss = []

# Compute WCSS for each k
for k in k_values:
    kmeans = KMeans(k)
    kmeans.fit(df.to_numpy())
    wcss.append(kmeans.wcss)
    print(f'k:{len(wcss)}   iter:{kmeans.num_iter}    wcss:{wcss[-1]}')

# Plot WCSS vs. k
plt.figure(figsize=(8, 5))
plt.plot(k_values, wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(True)
plt.show()
```

# According to Graph considering Elbow point **k=4**

```python
kmeans=KMeans(4)
cluster_assignments=kmeans.fit(df.to_numpy())
print(f'k:{kmeans.k}   iter:{kmeans.num_iter}    wcss:{kmeans.wcss}')
```

```python
for i in range(kmeans.k):
    print(f"Cluster {i+1}:")
    print(kmeans.centroids[i])
    print("\n")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(df.to_numpy())

# Plot the results
plt.figure(figsize=(8, 6))
colors = ['#4169E1', '#DC143C', '#32CD32', '#FFD700']
for cluster in range(kmeans.k):
    cluster_points = X_reduced[cluster_assignments == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f"Cluster {cluster}",color=colors[cluster])

# Plot centroids (projected to 2D)
centroids_reduced = pca.transform(kmeans.centroids)
plt.scatter(centroids_reduced[:, 0], centroids_reduced[:, 1], c='black', marker='X', s=100, label='Centroids')

plt.title("KMeans Clustering (2D PCA Projection)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```
